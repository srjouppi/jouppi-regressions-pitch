# Regressions

```python
### ODD That a negative relationship shows up -- the longer the permit, the lower the rate of fce_per_year
import statsmodels.formula.api as smf
model = smf.ols("fce_per_year ~ years_converted", data=df)
results = model.fit()
results.summary()
```

```python
# pd.qcut(df.years_converted,4)
```

```python
(
    ggplot(df.query('years_converted.notna()'), aes(x='years')) + 
        geom_histogram( bins=9) +
        facet_wrap('fce + sar', scales='free_y') + 
        theme(figure_size=(16,9))
)
```

```python
(
    ggplot(df, aes(x='years', y='fce + sar')) + 
        geom_point(alpha=.2, size=4) +
        theme(figure_size=(16,9))
)

# pd.crosstab(df.years_converted.quantile(9), df.fce)
    
```

```python
model = smf.ols("fce ~ sources", data=df_zip)
results = model.fit()
results.summary()
```

```python
df_zip.assign(residuals = results.resid,
             residuals_z = results.resid / results.resid.std()
).sort_values(by='residuals').query('fce.notnull()')
```

```python
model = smf.ols("sources ~ black_pct", data=df_zip)
results = model.fit()
results.summary()
```
